```python
import pandas as pd

# File to Load (Remember to change the path if needed.)
vine_data_load = "vine_table.csv"
vine_data_df = pd.read_csv(vine_data_load)
```

```python
vine_data_df
```

```python
total_votes_df = vine_data_df[vine_data_df["total_votes"] >= 20]
total_votes_df
```

```python
helpful_votes_df = total_votes_df[total_votes_df["helpful_votes"]/total_votes_df["total_votes"] >= 0.5]
helpful_votes_df
```

```python
paid_votes_df = helpful_votes_df[helpful_votes_df["vine"] == 'Y']
paid_votes_df
```

```python
unpaid_votes_df = helpful_votes_df[helpful_votes_df["vine"] == 'N']
unpaid_votes_df
```

```python
#total paid reviews
paid_votes_df.count()
```

```python
#paid 5 star reviews and percentage of 5 star reviews
paid_five_star_votes_df = paid_votes_df[paid_votes_df["star_rating"] == 5]
paid_five_star_votes_df
```

```python
# Percentage of paid rating
percentage_of_paid_rating_df = 100*paid_five_star_votes_df.count()["review_id"]/paid_votes_df.count()["review_id"]
percentage_of_paid_rating_df
```

```python
#total unpaid reviews
unpaid_votes_df.count()
```

```python
#unpaid 5 star reviews and percentage of 5 star reviews
unpaid_five_star_votes_df = unpaid_votes_df[unpaid_votes_df["star_rating"] == 5]
unpaid_five_star_votes_df
```

```python
# Percentage of unpaid rating
percentage_of_unpaid_rating_df = 100*unpaid_five_star_votes_df.count()["review_id"]/unpaid_votes_df.count()["review_id"]
percentage_of_unpaid_rating_df
```
